```python
import pandas as pd
```

```python
df = pd.read_csv('best sellin books total.csv', encoding='latin-1', on_bad_lines='skip')
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
df['price'].unique()[:20]
```

```python
#print("Dollar ($):", df['price'].str.contains(r'^\$').sum())
#print("Pound (Â£):", df['price'].str.contains(r'^Â£').sum())


df['currency_symbol'] = df['price'].str.extract(r'(^\D+)')
df['currency_symbol'].value_counts()
```

```python
df['price'] = df['price'].str.replace('$','', regex=False)
df['price'] = pd.to_numeric(df['price'], errors = 'coerce')
df['price'].describe()
#df.head()
```
